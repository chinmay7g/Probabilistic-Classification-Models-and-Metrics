```python
# Log Loss function / Binary Entropy function
```

```python
#importing libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
p = np.linspace( 1E-6, 1-1E-6, 1000)
y = 1

log_loss = -(y*np.log(p) + (1 - y)* np.log(1-p))
```

```python
# plotting results
plt.figure(figsize = [14,6])
plt.plot(p, log_loss)
plt.xlabel('probability')
plt.ylabel('Log_Loss')
plt.legend(['$y$ = 1'])
```
