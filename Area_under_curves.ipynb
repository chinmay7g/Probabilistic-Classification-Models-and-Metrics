```python
#importing libraries

import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
#generate data

np.random.seed(0)

y_prob = np.linspace(0,1,1000)
y_pred = (y_prob > 0.5).astype(np.int32)
y = np.random.binomial(1, y_prob)
```

```python
#getting accuracy of data applied above

accuracy = metrics.accuracy_score(y, y_pred)

print(f'The accuracy for the given data = {accuracy}')
```

```python
#dealing with precision and recall

precision, recall, threshold = metrics.precision_recall_curve(y, y_prob)

f1_score = 2*precision / ( precision + recall)

threshold = np.hstack((0, threshold))
```

```python
#visualizing all plots
plt.figure(figsize = [14,6])
plt.plot(threshold, precision)
plt.plot(threshold, recall)
plt.plot(threshold, f1_score)

plt.xlabel('Threshold')
plt.legend(['Precision','Recall', '$F_1$'])
```

```python
#visualising the graph for Precision vs Recall

plt.figure(figsize=[14,6])
plt.plot(recall, precision)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.xlim([0,1])
plt.ylim([0,1])
```

```python
#using scikit learn for AUC and ROC_AUC

auc = metrics.auc(recall, precision)
roc_auc = metrics.roc_auc_score(y, y_prob)
```

```python
print(f"The Area under curve for precision-recall = {auc}")
print(f'THe receiver-operator area under curve for y vs y_prob = {roc_auc}')
```
